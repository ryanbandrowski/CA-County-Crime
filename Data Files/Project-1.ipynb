```python
 %matplotlib notebook
import pandas as pd
 import matplotlib.pyplot as plt
import numpy as np
```

```python
output_data = pd.read_csv('Output Data Clean.csv')
output_data.head()
```

```python
# Filter the DataFrame down only to those columns to chart
population = output_data[["County","Population"]]

# Set the index to be "State" so they will be used as labels
population = population.set_index("County")

population.head()
```

```python
population.plot(kind="bar", figsize=(40,6))

# Set a title for the chart
plt.title("Population by California County")

plt.show()
plt.tight_layout()
```
