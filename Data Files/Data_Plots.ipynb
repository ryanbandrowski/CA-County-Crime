```python
 %matplotlib notebook
import pandas as pd
 import matplotlib.pyplot as plt
import numpy as np 

from scipy import stats


```

```python
output_data = pd.read_csv('Output Data Clean.csv')
output_data.head()

output_data['Population Density'] = output_data['Population']/output_data['Land Area']

output_data['Total Crime'] = output_data['Violent Crime'] + output_data['Property Crime']
output_data['Citizens per Officer'] = output_data['Population']/output_data['Total Officers']


output_data.sort_values(by='Total Crime', ascending=True)
```

```python
violent_crime_pop = output_data[['Violent Crime', 'Population Density']]
lines = violent_crime_pop.plot.scatter(x='Population Density', y='Violent Crime') 

```

```python
property_crime_pop = output_data[['Property Crime' , 'Population Density']]
lines = property_crime_pop.plot.scatter(x='Population Density', y='Property Crime')
```

```python
officers_pop = output_data[['Total Officers', 'Population']]
lines2 = officers_pop.plot.scatter(x='Total Officers', y='Population') 
```

```python
violent_pop = output_data[['Violent Crime', 'Population']]
lines3 = violent_pop.plot.scatter(x='Population', y='Violent Crime') 
```

```python
property_pop = output_data[['Property Crime', 'Population']]
lines6 = property_pop.plot.scatter(x='Population', y='Property Crime') 
```

```python
per_officer_crime = output_data[['Total Crime', 'Citizens per Officer']]
lines4 = per_officer_crime.plot.scatter(x='Citizens per Officer', y='Total Crime') 
```
