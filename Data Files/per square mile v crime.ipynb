```python
 %matplotlib notebook
import pandas as pd
 import matplotlib.pyplot as plt
import numpy as np 
from scipy import stats
```

```python
output_data = pd.read_csv('Output Data Clean.csv')
output_data.head()
output_data['Population Density'] = output_data['Population']/output_data['Land Area']
output_data['Total Crime'] = output_data['Violent Crime'] + output_data['Property Crime']
output_data['Citizens per Officer'] = output_data['Population']/output_data['Total Officers']
output_data['Officer per Square Mile'] = output_data['Total Officers'] / output_data['Land Area']
output_data['Total Crime Rate'] = output_data['Total Crime'] / output_data['Population'] * 100000
output_data['Violent Crime Rate'] = output_data['Violent Crime'] / output_data['Population'] * 100000
output_data['Property Crime Rate'] = output_data['Property Crime'] / output_data['Population'] * 100000
output_data.sort_values(by='Violent Crime', ascending=False).head()
output_data = output_data.drop([21])
output_data = output_data.drop([36])
output_data.head()
output_data.sort_values(by='Total Crime Rate', ascending=False).head()
```

```python
sqaure_mile = output_data.iloc[:,17]
```

```python
total_crime = output_data.iloc[:, 18]
tc_slope, tc_int, tc_r, tc_p, tc_std_err = stats.linregress(
   sqaure_mile, total_crime) 
tc_fit = tc_slope * sqaure_mile + tc_int
```

```python
violent_crime = output_data.iloc[:, 19]
vc_slope, vc_int, vc_r, vc_p, vc_std_err = stats.linregress(
   sqaure_mile, violent_crime) 
vc_fit = vc_slope * sqaure_mile + vc_int
```

```python
property_crime = output_data.iloc[:, 20]
pc_slope, pc_int, pc_r, pc_p, pc_std_err = stats.linregress(
   sqaure_mile, property_crime) 
pc_fit = pc_slope * sqaure_mile + pc_int
```

```python
 # Plot
fig, (ax1, ax2, ax3) = plt.subplots(3, sharex=True)
fig.suptitle("Officers per Square Mile v Crime Rate", fontsize=16, fontweight="bold")

ax1.set_xlim(min(sqaure_mile), max(sqaure_mile))
ax1.plot(sqaure_mile, total_crime, linewidth=0, marker="o")
ax1.plot(sqaure_mile, tc_fit, "b--", linewidth=1)
ax1.set_ylabel("Total Crime Rate")

ax2.plot(sqaure_mile,violent_crime , linewidth=0, marker="o", color="r")
ax2.plot(sqaure_mile, vc_fit, "r--", linewidth=1)
ax2.set_ylabel("Violent Crime Rate")

ax3.plot(sqaure_mile,property_crime , linewidth=0, marker="o", color="g")
ax3.plot(sqaure_mile, pc_fit, "g--", linewidth=1)
ax3.set_ylabel("Property Crime Rate")
ax3.set_xlabel("Officers Per Square Mile") 
```
