```python
 %matplotlib notebook
import pandas as pd
 import matplotlib.pyplot as plt
import numpy as np 
from scipy import stats
```

```python
output_data = pd.read_csv('Output Data Clean.csv')
output_data.head()
output_data['Population Density'] = output_data['Population']/output_data['Land Area']
output_data['Total Crime'] = output_data['Violent Crime'] + output_data['Property Crime']
output_data['Citizens per Officer'] = output_data['Population']/output_data['Total Officers']
output_data['Officer per Square Mile'] = output_data['Total Officers'] / output_data['Land Area']
output_data['Total Crime Rate'] = output_data['Total Crime'] / output_data['Population'] * 100000
output_data['Violent Crime Rate'] = output_data['Violent Crime'] / output_data['Population'] * 100000
output_data['Property Crime Rate'] = output_data['Property Crime'] / output_data['Population'] * 100000
output_data.sort_values(by='Violent Crime', ascending=False).head()
output_data = output_data.drop([21])
output_data = output_data.drop([36])
output_data.head()
output_data.sort_values(by='Total Crime Rate', ascending=False).head()
```

```python
#output_data['County'] = output_data['County'].str.replace(',', '').astype(int)
population = output_data.iloc[:,12]
```

```python
violent_crime_rate = output_data.iloc[:, 19]
#violent_crime_rate
vc_slope, vc_int, vc_r, vc_p, vc_std_err = stats.linregress(
   population, violent_crime_rate) 
vc_fit = vc_slope * population + vc_int
```

```python
property_crime_rate = output_data.iloc[:, 20]
pc_slope, pc_int, pc_r, pc_p, pc_std_err = stats.linregress(population,property_crime_rate )
pc_fit = pc_slope * population + pc_int
```

```python
total_crime = output_data.iloc[:,18]
tc_slope, tc_int, tc_r, tc_p, tc_std_err = stats.linregress(population,total_crime )
tc_fit = tc_slope * population + tc_int
```

```python
 # Plot
fig, (ax1, ax2, ax3) = plt.subplots(3, sharex=True)
fig.suptitle("Crime Rates vs Total Population", fontsize=16, fontweight="bold")

ax1.set_xlim(min(population), max(population))
ax1.plot(population, violent_crime_rate, linewidth=0, marker="o")
ax1.plot(population, vc_fit, "b--", linewidth=1)
ax1.set_ylabel(" Violent Crime Rate")

ax2.plot(population,property_crime_rate , linewidth=0, marker="o", color="r")
ax2.plot(population, pc_fit, "r--", linewidth=1)
ax2.set_ylabel("Property Crime Rate")

ax3.plot(population,total_crime , linewidth=0, marker="o", color="g")
ax3.plot(population, tc_fit, "g--", linewidth=1)
ax3.set_ylabel("Total Crime Rate")
ax3.set_xlabel("Population")
```
